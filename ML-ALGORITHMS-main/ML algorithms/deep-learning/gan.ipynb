```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten, Reshape
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

# Load and prepare the MNIST dataset
(X_train, _), (_, _) = tf.keras.datasets.mnist.load_data()
X_train = X_train.astype(np.float32) / 255.0
X_train = np.expand_dims(X_train, axis=-1)

# Define the discriminator model
def build_discriminator(image_shape):
    model = Sequential([
        Flatten(input_shape=image_shape),
        Dense(150, activation='relu'),
        Dense(100, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    return model

# Define the generator model
def build_generator(latent_dim):
    model = Sequential([
        Dense(128, activation='relu', input_dim=latent_dim),
        Dense(784, activation='sigmoid'),
        Reshape((28, 28, 1))
    ])
    return model

# Define the GAN model combining the discriminator and generator
def build_gan(generator, discriminator):
    model = Sequential([
        generator,
        discriminator
    ])
    return model

# Model settings
image_shape = (28, 28, 1)
latent_dim = 100

# Build and compile the discriminator
discriminator = build_discriminator(image_shape)
discriminator.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['accuracy'])

# Build the generator
generator = build_generator(latent_dim)

# Keep discriminator's parameters fixed when training gan
discriminator.trainable = False

# Build and compile the gan
gan = build_gan(generator, discriminator)
gan.compile(loss='binary_crossentropy', optimizer=Adam())

# Function to train the GAN
def train_gan(gan, generator, discriminator, dataset, latent_dim, epochs=100, batch_size=128):
    for epoch in range(epochs):
        # Random inputs for the generator
        random_latent_vectors = np.random.normal(0, 1, size=(batch_size, latent_dim))
        
        # Generate fake images
        generated_images = generator.predict(random_latent_vectors)
        
        indices = np.random.randint(0, dataset.shape[0], size=batch_size)
        # Get real images using these indices
        real_images = dataset[indices]
        
        # Labels for generated and real data
        generated_labels = np.zeros((batch_size, 1))
        real_labels = np.ones((batch_size, 1))
        
        # Train the discriminator
        d_loss_real = discriminator.train_on_batch(real_images, real_labels)
        d_loss_fake = discriminator.train_on_batch(generated_images, generated_labels)
        
        # Labels for GAN (trick to the discriminator)
        misleading_labels = np.ones((batch_size, 1))
        
        # Train the generator via GAN
        g_loss = gan.train_on_batch(random_latent_vectors, misleading_labels)
        
        # Print status
        print(f"Epoch: {epoch + 1}, D Loss Real: {d_loss_real[0]}, D Loss Fake: {d_loss_fake[0]}, G Loss: {g_loss}")

        # Optionally plot generated images per epoch
        if epoch % 10 == 0:
            plot_generated_images(epoch, generator)

# Function to display generated images
def plot_generated_images(epoch, generator, examples=10, dim=(1, 10), figsize=(10, 1)):
    noise = np.random.normal(0, 1, size=[examples, latent_dim])
    generated_images = generator.predict(noise)
    plt.figure(figsize=figsize)
    for i in range(examples):
        plt.subplot(dim[0], dim[1], i+1)
        plt.imshow(generated_images[i, :, :, 0], interpolation='nearest', cmap='gray')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Train the GAN
train_gan(gan, generator, discriminator, X_train, latent_dim, epochs=100)
```
