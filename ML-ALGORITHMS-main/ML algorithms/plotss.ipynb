```python
import pandas as pd


data = {
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sales': [200, 300, 400, 500, 600],
    'Profit': [20, 50, 60, 90, 110],
    'Category': ['A', 'B', 'C', 'D', 'E']
}

df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

# Line Plot Example
plt.plot(df['Year'], df['Sales'], label='Sales Over Time')
plt.title('Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Bar Plot Example
plt.bar(df['Category'], df['Profit'], color='skyblue')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()

# Scatter Plot Example
plt.scatter(df['Sales'], df['Profit'], color='green')
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()

# Histogram Example
plt.hist(df['Sales'], bins=5, color='purple')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()

# Pie Chart Example
plt.pie(df['Sales'], labels=df['Category'], autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution by Category')
plt.axis('equal')  # Ensures that pie is drawn as a circle.
plt.show()
```
