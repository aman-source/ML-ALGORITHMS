```python
def calculate_pure_alcohol(drinks_df):
    drinks_df['total_litres_of_pure_alcohol'] = drinks_df['beer_servings'].astype(float)*5+drinks_df['spirit_servings'].astype(float)*40+drinks_df['wine_servings'].astype(float)*15
    return drinks_df
    
drinks_df = calculate_pure_alcohol(drinks_df)
```

```python
import pandas as pd

sales_data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Units_Sold': [150, 250, 300, 100],
    'Price_per_Unit': [20, 15, 30, 10],
    'Discount': [0.1, 0.05, 0.2, 0.15]
}

sales_df = pd.DataFrame(sales_data)
print(sales_df)
```

```python
def total_revenue(row):
    return (row['Units_Sold']*row['Price_per_Unit'])*(1-row['Discount'])
    

sales_df['Total_Revenue'] = sales_df.apply(total_revenue,axis=1)
sales_df
```
